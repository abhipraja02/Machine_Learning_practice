```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import MinMaxScaler
import seaborn as sn
```

```python
df=pd.read_csv('D:/abhipraja/codes/dataset/breastCancer.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
train,test=train_test_split(df,random_state=42)
xtrain=train[train.columns[2:-1]]
ytrain=train['diagnosis']
xtest=test[test.columns[2:-1]]
ytest=test['diagnosis']
```

```python
scaler=MinMaxScaler()
xtrainScaled=scaler.fit_transform(xtrain)
xtestScaled=scaler.transform(xtest)
svcModel=SVC(random_state=0).fit(xtrainScaled,ytrain)
print("train score - " + str(svcModel.score(xtrainScaled, ytrain)))
print("test score - " + str(svcModel.score(xtestScaled, ytest)))
```

```python
print(svcModel)
```

```python
params={'C':[0.001,0.01,.1,1,10,100],
'gamma':[0.001,0.01,0.1,1,10,100]}
gridSearch=GridSearchCV(SVC(random_state=0),params,cv=5,n_jobs=-1)
gridSearch.fit(xtrain,ytrain)
print("train score - " + str(svcModel.score(xtrainScaled, ytrain)))
print("test score - " + str(svcModel.score(xtestScaled, ytest)))
```

```python
print(gridSearch.best_params_)
```

```python
resultsDF=pd.DataFrame(gridSearch.cv_results_)
scores=np.array(resultsDF.mean_test_score).reshape(6,6)
```

```python
resultsDF.head()
```

```python
sn.heatmap(scores,annot=True,xticklabels=params['gamma'],yticklabels=params['C'])
```
